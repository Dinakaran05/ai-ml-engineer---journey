```python
import asyncio
import time

async def call_llm_api(request_id: str, delay: int):
  """Simulates an LLM API call that takes time to respond."""
  print(f"[{request_id}] Sending request to LLM...")
  await asyncio.sleep(delay)
  print(f"[{request_id}] Response received after {delay}s")
  return f"Response for {request_id}"

async def main():
  start = time.time()

  results = await asyncio.gather(
      call_llm_api("Request-A", 2),
      call_llm_api("Request-B", 3),
      call_llm_api("Request-C", 1),
  )

  print(results)
  print(f"total time taken: {time.time() - start:.2f}s")

await main()
```
