```python
class MLModel:
  def __init__(self, model_name: str, learning_rate: float = 0.01):
    self.model_name = model_name
    self.learning_rate = learning_rate
    self.is_trained = False

  def train(self, data):
    print(f"Training '{self.model_name}' with learning_rate={self.learning_rate}...")
    self.is_trained = True
    print(f"Training complete. Data points used: {len(data)}")

  def predict(self, input_value):
    if not self.is_trained:
      raise ValueError(f"{self.model_name} is not trained yet!")
    return input_value * self.learning_rate

model_a = MLModel(model_name="HousePricePredictor", learning_rate=0.05)
model_b = MLModel(model_name="SpamClassifier", learning_rate=0.1)

model_a.train(data=[1, 2, 3, 4, 5])
print(model_a.predict(100))

print(model_b.is_trained)



```

```python
class BankAccount:
  def __init__(self, account_holder, balance=0):
    self.account_holder = account_holder
    self.balance = balance

  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
    else:
      print("Deposit amount must be positive.")

  def withdraw(self, amount):
    if amount <= self.balance:
      self.balance -= amount
    else:
      print("Insufficient balance")

  def check_balance(self):
    print(f"Current balance: {self.balance}")

account1 = BankAccount("Dinakaran", 1000)
account2 = BankAccount("Karthik", 500)

account1.deposit(500)
account1.withdraw(200)
account1.check_balance()

account2.withdraw(1000)
account2.check_balance()

```
